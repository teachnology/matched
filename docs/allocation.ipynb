# Allocation algorithm

In this tutorial, we will explore the allocation algorithm. Before we can run the algorithm, we will first introduce the data.

### `projects.csv`

```python
import pathlib
import pandas as pd

raw = pathlib.Path("./raw")

projects = pd.read_csv(raw / "projects.csv").set_index("code")

projects.head(3)
```

`projects.csv` contains some of the data we collected from supervisors. Each project is given a unique code (`code (str)`). For each project code, we record:
- `external (bool)` indicating the project is external or internal
- `nmax (int>0)` - the maximum number of students that can be assigned to the project
- `course_i (bool)` - what course(s) the project is available for.

From the data above, we can see that the project `mabe-007` is an internal project that can be assigned to at most 3 students, and is only available for students on course1.
